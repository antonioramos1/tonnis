Group slams & rank by difficulty

```python
#https://github.com/mwaskom/seaborn/issues/1076
from matplotlib.patches import PathPatch
def adjust_box_widths(g, fac):
    """
    Adjust the widths of a seaborn-generated boxplot.
    """

    # iterating through Axes instances
    for ax in g.axes:

        # iterating through axes artists:
        for c in ax.get_children():

            # searching for PathPatches
            if isinstance(c, PathPatch):
                # getting current width of box:
                p = c.get_path()
                verts = p.vertices
                verts_sub = verts[:-1]
                xmin = np.min(verts_sub[:, 0])
                xmax = np.max(verts_sub[:, 0])
                xmid = 0.5*(xmin+xmax)
                xhalf = 0.5*(xmax - xmin)

                # setting new width of box
                xmin_new = xmid-fac*xhalf
                xmax_new = xmid+fac*xhalf
                verts_sub[verts_sub[:, 0] == xmin, 0] = xmin_new
                verts_sub[verts_sub[:, 0] == xmax, 0] = xmax_new

                # setting new width of median line
                for l in ax.lines:
                    if np.all(l.get_xdata() == [xmin, xmax]):
                        l.set_xdata([xmin_new, xmax_new])

#https://stackoverflow.com/questions/22566284/matplotlib-how-to-plot-images-instead-of-points
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

def imscatter(x, y, image, ax=None, zoom=1):
    """Use custom images as points"""
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
```

```python
dpi = 600
transparent = True
ylim = (0,2.1)
xlim = (-10, 110)
figsize = (15,25)

plot_edit_params = {'axes.facecolor':'black', 'figure.facecolor':'black',
            "axes.spines.left": True, "axes.spines.right": False, "axes.spines.top": False, "axes.spines.bottom": False,
            "xtick.bottom":False, "xtick.top":False, "ytick.left": False, "ytick.right": False, "ytick.labelleft": True,
            "axes.grid": False,
            "text.color": "white", "axes.labelcolor":"white", "axes.edgecolor": "white", "ytick.color": "white",
            "font.size": 11, "font.family":"DejaVu Sans Mono", "axes.labelsize": 15, "ytick.labelsize": 20, "axes.titlesize": 15,
           }
sns.set(rc=plot_edit_params)
```
